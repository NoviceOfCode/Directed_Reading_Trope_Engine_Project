# Data Retrieval

```python
df = pd.read_csv('../data_clean/parsed_tropes.csv')

df.head()
```

```python
nlp = spacy.load("en_core_web_sm")
stop_words = set(stopwords.words('english'))
```

```python
def preprocess(text_lst: List[str]):
    text_clean = []
    text_loss = []
    text_loss_scores = []
    for text in text_lst:
        # Lowercasing words
        text = text.lower()
        # Pass to lemmatizer
        doc = nlp(text)
        tokens = []
        token_loss = []
        token_loss_score = 0

        # process each token
        for token in doc:
            if token.dep_ == 'neg':
                head = token.head.lemma_
                neg_token = 'not_' + head
                tokens.append(neg_token)
            elif token.lemma_.lower() not in stop_words and token.is_alpha and detect(token.lemma_.lower()) != 'ja':
                tokens.append(token.lemma_)
            elif token.is_alpha:
              token_loss.append(token.lemma_)
              token_loss_score += 1
        text_loss_scores.append(token_loss_score)
        text_loss.append(token_loss)
        text_clean.append(tokens)
    return text_clean, text_loss, text_loss_scores
```

```python
df['mha_example_usage'] = df['mha_example_usage'].apply(ast.literal_eval)
```

```python
df_copy = df.copy()
df_copy[['mha_example_usage', 'text_loss','text_loss_scores']] = df_copy['mha_example_usage'].apply(preprocess).apply(pd.Series)
```
